20. Explain the difference between a list and a dictionary in terms of data retrieval.
  - the key difference between a list and a dictionary when it comes to data retrieval lies in how you access elements and what you're accessing.
  - Position-Based Access:
    - Element in a list are accessed using their index, which reflect their position in the sequence.
    - Retrieval is order sensitive and requires knowing the element's position.
  - **Dictionaries:key-Based Access:**
    - Element in a dictionary are accessed using a key.
     - Retrieval is label-sensitive and order isn't necessary.
